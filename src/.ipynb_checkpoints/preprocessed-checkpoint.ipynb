```python
data.isna().sum()
```

```python
data.drop(['snapshot_radar', 'metaf'], axis = 1, inplace = True)
```

```python
data.dropna(inplace = True, axis = 0)
```

```python
data.isna().sum()
```

```python
data['wind_variation'] = None
data['qnh'] = None
data["cavok"] = None
data["gust"] = None
data["OVC_presenceCB"] = 0.0
data["SCT_presenceCB"] = 0.0
data["FEW_presenceCB"] = 0.0
data["BKN_presenceCB"] = 0.0
data["OVC_presenceTCU"] = 0.0
data["SCT_presenceTCU"] = 0.0
data["FEW_presenceTCU"] = 0.0
data["BKN_presenceTCU"] = 0.0
```

```python
for index, row in data.iterrows():
    translationMetar = Metar(row['destino'], row['metar'])
    data.loc[index, 'cavok'] = float(bool(re.search('CAVOK', row['metar'])))

    if(translationMetar.wind):
        data.loc[index, 'wind_variation'] = float(translationMetar.wind['variation'] != None)
        data.loc[index, 'gust'] = float(translationMetar.wind['gust'] != None)
    else:
        # Se não tem informação de vento significa METAR no formato ex.: VRB07G17KT (vento variavel com rajada)
        data.loc[index, 'gust'] = 1


    if(translationMetar.qnh):
        data.loc[index, 'qnh'] = float(translationMetar.qnh)

    if(translationMetar.cloud):
        for formation in translationMetar.cloud:
            data.loc[index, f"{formation['code']}_presenceCB"] = float(formation['presenceCB']) 
            data.loc[index, f"{formation['code']}_presenceTCU"] = float(formation['presenceTCU'])
```

```python
data.isna().sum()
```

```python
data.dropna(inplace = True, axis = 0)
```
